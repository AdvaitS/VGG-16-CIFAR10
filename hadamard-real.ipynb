```python
from sympy import fwht, ifwht
from matplotlib import pyplot
import numpy as np
df2 = unpickle("/Users/advaitshukla/Downloads/cifar-10-batches-py/data_batch_1")
alldata = df2[b'data']
alldata = alldata.reshape(len(alldata),3,32,32).transpose(0,2,3,1)
pic1 = alldata[7]
pyplot.subplot(5,1,1)
pyplot.title("Original Image")
pyplot.imshow(pic1)
h_transformed = np.array(fwht(pic1), dtype=int)
print(h_transformed.shape)
pyplot.subplot(5,1,3)
pyplot.title("Hadamard transformed")
pyplot.imshow(h_transformed)
ih_transformed = np.array(ifwht(h_transformed), dtype=int)
pyplot.subplot(5,1,5)
pyplot.title("Inverse Hadamard transformed")
pyplot.imshow(ih_transformed)
```
